```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
%matplotlib
```

```python
data=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
```

```python
data.head()
```

```python
### cleaning he dataset
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data[data.isnull().any(axis=1)]
```

```python
data.loc[:122,"Region"]=0
data.loc[122:,"Region"]=1
data.info()

```

```python
data.head()
```

```python
data[["Region"]]=data[["Region"]].astype(int)
data.info()
```

```python
data.isnull().sum()
```

```python
data[data.isnull().any(axis=1)]
```

```python
#removing null values
df=data
df=df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
#check the rows
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
df.iloc[[122]]
```

```python
#fixing the spaces in columns names
df.columns=df.columns.str.strip()
df.columns
```

### Changing the required columns as integer datatype

```python
df[['month','year','Temperature','RH','Ws','day']]=df[['month','year','Temperature','RH','Ws','day']].astype(int)
```

```python
df.info()
```

### changing the other columns to float



```python
objects=[features for features in df.columns if df[features].dtypes=='O' ]
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

### saving the cleaned dataset

```python
df.to_csv('Algerian_forest_fires_clean.csv',index=False)
```
